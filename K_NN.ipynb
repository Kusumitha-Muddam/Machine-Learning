**Student Placement Prediction With KNN**

```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import confusion_matrix,classification_report
```

```python
df=pd.read_csv('placement.csv')
df.head()
```

```python
df.info()
```

```python
print(df['PlacementStatus'].value_counts())
```

```python
# target
y=df["PlacementStatus"]

# features
x=df.drop(columns=["PlacementStatus"])

x=x.drop(columns=["StudentID"])
```

```python
x.head(2)
```

```python
#num_x=x.select_dtypes(include=np.number)
cat_x=x.select_dtypes(exclude=np.number)
cat_x
```

one-hot Encoding

```python
x_enc=pd.get_dummies(cat_x)
x_enc
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_enc=le.fit_transform(y)
y_enc
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_enc,y_enc,test_size=0.2,random_state=2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#SCALING
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
#MODEL

knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled,y_train)
y_pred=knn.predict(x_test_scaled)
```

```python
#evaluation
acc=accuracy_score(y_test,y_pred)
print(acc)
cm=confusion_matrix(y_test,y_pred)
print(cm)
cr=classification_report(y_test,y_pred)
print(cr)
```
